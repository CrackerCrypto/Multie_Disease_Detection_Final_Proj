```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('D:\\Project_Disease_Prediction\\datasets\\fetal_health.csv')
df.head()
```

```python
val = df.isnull().values.any()
print(val)
```

```python
df = df.rename(columns = {'baseline value':'baseline_value', 'prolongued_decelerations': 'prolonged_decelerations'})
```

```python
df['fetal_health'].unique()
```

```python
df['fetal_health'] = df['fetal_health'].replace([2.0, 1.0, 3.0], [1.0, 0.0, 2.0])
df.head()
```

```python
df['fetal_health'].unique()
```

# STARTING ANN MODEL

```python
df.head()
```

```python
X = df.iloc[:, :21]
Y = df.loc[:,['fetal_health']]

print(X.shape)
print(Y.shape)
```

```python
no_input_nodes = X.shape[1]
print(no_input_nodes)

no_output_nodes = len(Y['fetal_health'].unique())
print(no_output_nodes)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 44)
```

```python
Classification_model = Sequential(
    [
        tf.keras.Input(shape=(no_input_nodes, )),
        Dense(256, activation='relu', name='layer_1'),
        Dropout(0.2),
        Dense(128, activation='relu', name='layer_2'),
        Dropout(0.2),
        Dense(64, activation='relu', name='layer_3'),
        Dropout(0.2),
        Dense(no_output_nodes, activation='softmax', name='output_layer')
    ]
)

Classification_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
```

```python
Classification_model.summary()
```

```python
Classification_model.fit(X_train, y_train, batch_size=1, epochs=50)
```

```python
score, acc = Classification_model.evaluate(X_train, y_train, batch_size=1)

print('Train score', score *100)
print('Train accuracy', acc *100)
```
